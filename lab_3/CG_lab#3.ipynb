```python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation
```

## Task 3-5

```python
def generate_fractal_sequence(axiom, rules, iterations):
    for _ in range(iterations):
        next_sequence = ""
        for symbol in axiom:
            next_sequence += rules.get(symbol, symbol)
        axiom = next_sequence
    return axiom

def draw_fractal(sequence, angle, distance):
    direction = np.pi / 2  
    x, y = 0, 0  
    path_x, path_y = [x], [y]  
    
    for command in sequence:
        if command == 'F':
            x += distance * np.cos(direction)
            y += distance * np.sin(direction)
            path_x.append(x)
            path_y.append(y)
        elif command == '+':
            direction += np.radians(angle) 
        elif command == '-':
            direction -= np.radians(angle)  
    
    return path_x, path_y 

def update(frame):
    angle, axiom, rules, iterations, distance = frame
    ax.clear()
    ax.axis('equal')
    ax.axis('off')
    sequence = generate_fractal_sequence(axiom, rules, iterations)
    path_x, path_y = draw_fractal(sequence, angle, distance)
    ax.plot(path_x, path_y, color='black')
    ax.set_title(f"Angle: {angle}°") 
```

### Хрестоподібний фрактал

```python
# fractal parameters
axiom = "F+XF+F+XF"
rules = {"X": "XF-F+F-XF+F+XF-F+F-X"}
iterations = 4  
angle = 90
distance = 5  

path_x, path_y = draw_fractal(generate_fractal_sequence(axiom, rules, iterations), angle, distance)

plt.figure(figsize=(10, 10))
plt.plot(path_x, path_y, color='black')
plt.axis('equal')
plt.axis('off')  
plt.show()
```

```python
fig, ax = plt.subplots()

frames = [(angle, axiom, rules, iterations, distance) for angle in np.linspace(0, 90, 91)]

ani = FuncAnimation(fig, update, frames=frames, repeat=False)

plt.show()
```
